```python
file_path = open("data/shakespeare.txt", 'r')

poems = file_path.readlines()
poems
```

```python
from itertools import groupby
split_at = "\n"
final_poems = [list(g)[1:] for k, g in groupby(poems, lambda x: x != split_at) if k]

final_poems = [''.join([line.strip(' ') for line in poem]) for poem in final_poems]
final_poems
```

```python
[HMM_helper.parse_observations(poem) for poem in final_poems]
```

```python
# Using rhyming scheme ababcdcdefefgg

poem_lengths = [len(poem) for poem in final_poems] 
bad_poems = np.where(np.array(poem_lengths)!= 14)

final_poems = [final_poems.pop(ind) for ind in bad_poems]
final_poems
```
